Q3. Load the “UCI_liver_cirrhosis.CSV” dataset, from the raw_data module in CANVAS, into Python (see the data dictionary at the bottom of this document). This is a dataset used for predicting “Liver problems” in patients. Perform the EDA analysis by:
I.	Summarizing each numerical column (e.g., min, max, mean)
II.	Displaying scatter plots of “Age”,”TB”,”DB” and “Alkphos” one pair at a time
III.	Showing box plots for columns “Age”,”TB”,”DB” and “Alkphos”						


```python
#Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
df = pd.read_csv('UCI_liver_cirrhosis.csv')
df.head()
```

1. Summarizing each numerical column (e.g., min, max, mean)

```python
#Summarizing the numerical columns
summary = df.describe()
print("Summary of numerical columns:")
print(summary)
```

2. Displaying scatter plots of “Age”, ”TB”, ”DB” and “Alkphos” one pair at a time

```python
#List of columns that are required to be scatter plotted
columns = ["Age", "TB", "DB", "Alkphos"]
```

```python
# Scatter plot
for i in range(len(columns)):
    for j in range(i + 1, len(columns)):
        plt.figure(figsize=(6, 6))
        sns.scatterplot(data=df, x=columns[i], y=columns[j])
        plt.title(f"Scatter Plot: {columns[i]} vs {columns[j]}")
        plt.xlabel(columns[i])
        plt.ylabel(columns[j])
        plt.tight_layout()
        plt.show()
```

3. Showing box plots for columns “Age”, ”TB”, ”DB” and “Alkphos”.	

```python
#Box plot for the mentioned columns
for col in columns:
    plt.figure(figsize=(6, 6))
    sns.boxplot(y=df[col])
    plt.title(f"Box Plot: {col}")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```
